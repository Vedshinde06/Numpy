```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

x = [ 1,2,3,4,5]
y = [2,4,6,8,10]
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Plot')
plt.plot(x, y, color='red', marker='o', linestyle='--')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories, values, color='orange')
plt.show()
```

```python
sizes = [30, 40, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

Function	Description
plt.plot()	Line plot
plt.scatter()	Scatter plot
plt.bar()	Bar chart
plt.hist()	Histogram
plt.pie()	Pie chart
plt.xlabel()	Add label to the X-axis
plt.ylabel()	Add label to the Y-axis
plt.title()	Add title to the plot
plt.legend()	Add a legend
plt.grid()	Add grid lines
plt.savefig()	Save the plot to a file
